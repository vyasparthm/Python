### List Methods

```python
#Length
print('Before pop: ',len(movies))
movies.pop() #Remove last
print('After Pop: ',len(movies))
movies.extend(['2011'])
print('After Extend: ',movies)
movies.insert(0,'Batman')
print('After Insert: ',movies)
movies.remove('Batman')
print('After Removal: ',movies)
```

```python
# Now lets try an put a condition

for each_item in movies:
    if isinstance(each_item,list):
        for nested_item in each_item:
            print(nested_item)
    else:
        print(each_item)
```

```python
# lets get a little better version of what we just did above

for each_item in movies:
    if isinstance(each_item,list):
        for nested_item in each_item:
            if isinstance(nested_item,list):
                for deep_item in nested_item:
                    print(deep_item)
            else:
                print(nested_item)
    else:
        print(each_item)
```

## Donâ€™t repeat code; create a function
Take a look at the code that youâ€™ve created so far, which (in an effort to save
you from having your brain explode) has already been amended to process yet
another nested list. Notice anything?



```python
# How about we write a fuction rather than writing complex loops?
def nested_lists(the_list):
    for each_item in the_list:
        if isinstance(each_item,list):
            nested_lists(each_item)
        else:
            print(each_item)
```

```python
nested_lists(movies)
# NOTE:  Python 3 defaults its recursion limit to 1,000
```
