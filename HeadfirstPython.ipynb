### List Methods

```python
#Length
print('Before pop: ',len(movies))
movies.pop() #Remove last
print('After Pop: ',len(movies))
movies.extend(['2011'])
print('After Extend: ',movies)
movies.insert(0,'Batman')
print('After Insert: ',movies)
movies.remove('Batman')
print('After Removal: ',movies)
```
